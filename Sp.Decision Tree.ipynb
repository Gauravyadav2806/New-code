### Splitting Target Variable

```python
# Splitting target variable
df_predictor = bank.iloc[:, bank.columns != 'TARGET']
target = bank.iloc[:, bank.columns == 'TARGET']
```

```python
# Getting the names of the features (all dependent variables)
df_predictor.columns
```

```python
# Getting the name of the target variable
target.columns
```

### Separating categorical and numerical variables from data frame for dummy coding on categorical column




```python
## Save all categorical columns in list
cat_columns = [col for col in df_predictor.columns.values if df_predictor[col].dtype == 'object']

# dataframe with categorical features
bank_cat = df_predictor[cat_columns]
#dataframe with numerical features
bank_num = df_predictor.drop(cat_columns, axis=1)
```

## Dummy Encoding

### Perform 'Dummy Coding' for all the categorical variables

```python
# Using pandas.get_dummies function to convert categorical variable into dummy/indicator variables.
bank_cat_dummies = pd.get_dummies(bank_cat,drop_first=True)
```

```python
# Using concat function we merge two dataframe for further analysis
df_predictor = pd.concat([bank_num, bank_cat_dummies], axis=1)
```

```python
# Getting the first-five rows of predictor dataframe
df_predictor.head()
```

```python
df_predictor.shape
```

```python
# Let us now split the dataset into train and test
# import train_test_split from sklearn,model_selection
from sklearn.model_selection import train_test_split
#This code will perform train, test on our independent and dependent variable
#Shape : (x,y) =, where x is number of observations and y is number of varaibles
X_train, X_test, Y_train, Y_test = train_test_split(df_predictor, target, test_size=0.30, \
                                                   random_state=0)
print("X_Train", X_train.shape)
print("X_Test", X_test.shape)
print("Y_Train", Y_train.shape)
print("Y_Test", Y_test.shape)
```

```python
### Building a Decision Tree Model using 'gini' solver and splitter 'best
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_columns = None

clf_DT = DecisionTreeClassifier(criterion='gini', splitter='best', max_depth=18,
                               min_samples_split=2, min_samples_leaf=1,
                               min_weight_fraction_leaf=0.0,
                               min_impurity_split=1e-07)
# criterion- The function to measure the quality of a split
# Supported criteria are "gini" for the gini impurity and 'entropy' for the information gain.
# Splitter-The strategy used to choose the split at each node
# Supported strategies are "best" to choose the best split and "random" to choose the best random split
# max-depth = The maximum depth of the tree
# min_samples_split- The minimum no of samples required to split an internal node
# min_samples_leaf - The minimum no of samples required  to be at a leaf node
# min_weight_fraction_leaf- The minimum weighted fraction of the sum total of weights (of all the input samples) required to be at a leaf node.
# samples have equal weight when sample_weight is not provided
# min_impurity_split_Threshold for early stopping in tree growth
# A node will split if its impurity is above the threshold, Otherwise it is a leaf.


# fitting the model
clf_DT.fit(X_train, Y_train)
# Predicting the model
y_pred_DT = clf_DT.predict(X_test)
```

### Plotting the ROC Curve for Decision Tree Classifier

```python
# Calculate the predicted probabilty on the test data
y_pred_proba = clf_DT.predict_proba(X_test)[::,1]
# Creating the roc curve on Y_test and y_pred_proba
fpr,tpr,_ = metrics.roc_curve(Y_test,y_pred_proba)
#Find area under roc curve
auc = metrics.roc_auc_score(Y_test,y_pred_proba)
#Using plt.plot, plot auc score on the label
plt.plot(fpr,tpr,label = 'data1, auc ='+str(auc))
# Print out label on the graph and loc value
plt.legend(loc = 4)
# Showing the graph
plt.show()
```

### Results

#### Displaying results associated with Decsion Tree Model in a Tabular Format.

```python
# Create a list of parameter
cols = ['Model', 'ROC Score', 'Precision Score', 'Recall Score', 'Accuracy Score', 'Kappa Score']

# Create a dataframe of the above list
models_report = pd.DataFrame(columns = cols)

# Create a result associated with the list mentioned above
tmpl = pd.Series({'Model':" Decision Tree with (ENTROPY & BEST) ",
                 'ROC Score': metrics.roc_auc_score(Y_test, y_pred_DT),
                 'Precision Score': metrics.precision_score(Y_test, y_pred_DT),
                 'Recall Score': metrics.recall_score(Y_test, y_pred_DT),
                 'Accuracy Score': metrics.accuracy_score(Y_test, y_pred_DT),
                 'Kappa Score': metrics.cohen_kappa_score(Y_test, y_pred_DT)})

# Below code will append the above tmpl series in model report dataframe
model_dt_report = models_report.append(tmpl, ignore_index = True)
model_dt_report
```

## Grid Search CV

### Grid search is the process of performing hyper parameter tuning in order to determine the optimal values for a given model. This is significant as the performance of the entire model is based on the hyper parameter values specified.

```python
# Set the parameters for grid search 
grid_search_parameters = {'criterion' : ['gini', 'entropy'],
                         'min_samples_split':[2],
                         'max_depth': [None,2,3],
                         'min_samples_leaf': [1,5],
                         'max_leaf_nodes':[None],}

# criterion- The function to measure the quality of a split
# Supported criteria are "gini" for the gini impurity and 'entropy' for the information gain.
# max-depth = The maximum depth of the tree
# min_samples_split- The minimum no of samples required to split an internal node
# min_samples_leaf - The minimum no of samples required  to be at a leaf node
# max_leaf_nodes- Grow a tree with max_leaf_nodes in best-first fashion
# Best nodes are defined as relative reduction in impurity. If None then unlimited number of leaf nodes.
```

## Building a train model after doing Grid Search

```python
# Import Grid Search CV from sklearn.model_selection
from sklearn.model_selection import GridSearchCV

# Create an instance of Decision Tree Classifier
classifier = DecisionTreeClassifier()

# Use Grid Search CV() , pass the values you have set for grid search
model_DT_Grid = GridSearchCV(classifier, grid_search_parameters,cv = 3)
# Creating a train model using Grid Search
model_DT_Grid.fit(X_train,Y_train)
```

```python
model_DT_Grid.best_params_
```

```python
# Performing predictions on the trained model after doing Grid search
test_pred = model_DT_Grid.predict(X_test)
```

```python
# Import Confusion Matrix from sklearn
from sklearn.metrics import confusion_matrix
# This code will create confuison matrix of predicted and actual values
confusion_matrix(Y_test,test_pred)
```
